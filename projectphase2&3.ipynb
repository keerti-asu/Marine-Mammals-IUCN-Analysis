```python
#creates family dataframe
import pandas as pd
testdata = pd.read_csv("C:/Users/keert/OneDrive/Desktop/keerti/module6/IUCNHierarchyAndExtinction.csv")
family = testdata[['family']].copy()
family = family ["family"].str.strip()
family = family.map(str.lower)
duplicate_family = family[family.duplicated()]
family = family.drop_duplicates()
family = family.rename({'family':'fName'})
```

```python
#creates genus dataframe
genus = testdata[['genus', 'family']].copy()
genus['genus'] = genus["genus"].str.strip()
genus['family'] = genus["family"].str.strip()
genus.columns = genus.columns.map(str.lower)
duplicate_genus = genus[genus.duplicated()]
genus = genus.drop_duplicates()
genus = genus.rename(columns={'genus':'gName', 'family':'fName'})
```

```python
#creates species dataframe
species = testdata[['taxonid', 'species', 'friendly_name', 'category', 'genus']].copy()
species.columns = species.columns.map(str.lower)
duplicate_species = species[species.duplicated()]
species = species.drop_duplicates()
species = species.rename(columns={'taxonid':'taxonID', 'species':'sName', 'friendly_name':'friendlyName', 'genus':'gName'})
```

```python
#creates spCountry dataframe
#SpCountry (taxonID, country)
testdata2 = pd.read_csv("C:/Users/keert/OneDrive/Desktop/keerti/module6/IUCNSpeciesCountryOccurances.csv")
spCountry = testdata2[['taxonid', 'countryoccurrencelookup_value']].copy()
spCountry.columns = spCountry.columns.map(str.lower)
duplicate_spCountry = spCountry[spCountry.duplicated()]
spCountry = spCountry.drop_duplicates()
spCountry = spCountry.rename(columns={'taxonid':'taxonID', 'countryoccurrencelookup_value':'country'})
```

```python
#saving the dataframes as csv files
family.to_csv('family.csv', index=False)
genus.to_csv('genus.csv', index=False)
species.to_csv('species.csv', index=False)
spCountry.to_csv('spCountry.csv', index=False)
```
